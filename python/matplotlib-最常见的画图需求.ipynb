```python
line = np.arange(-8,8,0.5)
```

```python
coordinate_x, coordinate_y = np.meshgrid(line, line)
```

```python
z = np.square( np.square(coordinate_x) + np.square(coordinate_y) )
```

```python
z == np.square( coordinate_x**2 + coordinate_y**2 )
# 快速判断全部为True的函数，等其它快速判断函数。 is all
```

# curve

```python
x = np.linspace(-5,5,50)
y1, y2 = np.cos(x), np.sin(x)


plt.figure(figsize=(10,5), dpi=80)
plt.plot(x, y1, color='b', linewidth=2, linestyle="--", label='cosine')
plt.plot(x, y2, color='r', linewidth=2, linestyle="--", label='sine')
plt.legend(loc='upper right')
## lengend
# best, upper right, upper left, lower left, lower right, right
# center left, center right, lower center, upper center, center

plt.xlim(-5.5, 5.5)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.ylim(y1.min()*1.1, y1.max()*1.1)
plt.yticks(np.linspace(y1.min(), y1.max(), 3, endpoint=True))

## moving spines
# we’ll discard the top and right by setting their color to none 
# and we’ll move the bottom and left ones to coordinate 0 in data 
# space coordinates.
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# Annotate some points
t = 2*np.pi/3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2, linestyle='--')
plt.scatter([t,], [np.cos(t),], lw=30, color='b')
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
    xy = (t,  np.cos(t)), xycoords='data', xytext=(-90, -50), 
    textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
    xy=(t, np.sin(t)), xycoords='data',
    xytext=(+10, +30), textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

print(list(ax.get_xticklabels()))
print(list(ax.get_yticklabels()))
for label in ax.get_xticklabels() + ax.get_yticklabels():
#     print('before: ', label)
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
#     print('after: ', label)
print(list(ax.get_xticklabels()))
print(list(ax.get_yticklabels()))
```

```python
x = np.linspace(-np.pi, np.pi, 50, endpoint=True)
y = np.sin(2 * x)

plt.figure(figsize=(10,5), dpi=80)
ax = plt.axes([0.025, 0.025, 0.95, 0.95])

# ax.xaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.3))
# ax.yaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.3))
ax.grid(which='major', axis='x', linewidth=0.75, linestyle='-', color='0.75')
# ax.grid(which='minor', axis='x', linewidth=0.25, linestyle='-', color='0.75')
ax.grid(which='major', axis='y', linewidth=0.75, linestyle='-', color='0.75')
# ax.grid(which='minor', axis='y', linewidth=0.25, linestyle='-', color='0.75')
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.plot(x, y + 1, color='blue', alpha=1.00)
plt.fill_between(x, 1, y+1, color='b', alpha=0.1)
plt.plot(x, y - 1, color='blue', alpha=1.00)
plt.fill_between(x, -1, y-1, y-1>-1, color='b', alpha=0.1)
plt.fill_between(x, -1, y-1, y-1<-1, color='r', alpha=0.1)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
```
